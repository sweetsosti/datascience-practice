# 1. What is Data Science?

 Data Science is the field that uses math, statistics, programming, and domain knowledge to extract useful insights and make decisions from data. Or,
 Data Science is an interdisciplinary field that combines statistics, computer science, and domain knowledge to analyze and interpret data, derive insights, and support decision-making. It involves the entire lifecycle of data — from collection and cleaning to analysis, visualization, and building predictive models using machine learning.
 Or, <b> in simple word</b>
 Data Science is the art of turning raw data into meaningful stories and solutions.

# 2. Key Component of Data Science

* <b>Data Collection</b> – Getting data from files, web, databases, sensors, etc.
* <b>Data Cleaning</b> – Fixing missing, incorrect, or messy data.
* <b> Data Analysis</b> – Understanding the data using statistics and logic.
* <b>Data Visualization</b> – Making graphs to show patterns (e.g., with Matplotlib or Seaborn).
* <b> Machine Learning</b> – Creating models that learn from data and make predictions.
* <b> Communication</b> – Explaining results to non-technical people through reports, dashboards, or presentations.

# 2.1 Data Collection

Data Collection is the first and most crucial step in the Data Science workflow. It involves gathering raw data from different sources to be processed and analyzed for insights.

<b>Types of Data Sources</b>
CSV/Excel files – Local or online spreadsheets.
Databases – SQL (MySQL, PostgreSQL) or NoSQL (MongoDB).
Web Data – Scraped using tools like BeautifulSoup, Selenium, or APIs.
Sensors/IoT Devices – Real-time data (e.g., temperature, movement).
Surveys/Manual Entry – Collected through forms or interviews.
Open Datasets – From sites like Kaggle, UCI Machine Learning Repository, or data.gov.

<b>Tools for Data Collection in Python</b>
pandas.read_csv() – Load CSV files.
requests – Access APIs.
BeautifulSoup, Selenium – Web scraping.
SQLAlchemy / sqlite3 – Query databases.

 <b>Sources to Collect Data From</b>

1. Open Data Platforms
Kaggle – Datasets for competitions, learning, and projects.
UCI Machine Learning Repository – Classic machine learning datasets.
Google Dataset Search – A search engine for datasets.
Data.gov – US government public datasets.
World Bank Open Data – Global financial and development data.
UN Data – International statistics from the UN.

# classifying data

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot for target variable
sns.countplot(x='class', data=data)
plt.title("Class Distribution (Edible vs Poisonous)")
plt.show()
class_counts = data['class'].value_counts()
print(class_counts)
```

<b> The code generates a bar plot to visualize the distribution of edible vs. poisonous mushrooms in the dataset.Where 3916 mushrooms are poisonous and 4208 mushrooms are poisonous ( It uses **seaborn's countplot** to display the count of each class.
