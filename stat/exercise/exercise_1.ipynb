```python
	
min_thresold, max_thresold = data.price.quantile([0.01,0.999])
min_thresold, max_thresold
```

```python
data[data.price<min_thresold]
```

```python
data[data.price<max_thresold]
```

```python
df2 = data[(data.price>min_thresold)&(data.price<max_thresold)]
df2.shape
```
