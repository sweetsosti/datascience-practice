# To found how many time person get transaction

```python
transaction_counts = data.groupby(['Beneficiary Name', 'Transaction Status']).size().reset_index(name='Count')
print(transaction_counts)
```

```python
total_transactions = data.groupby('Beneficiary Name').size().reset_index(name='TotalCount')
print(total_transactions)
```

```python

# Merge totals with breakdown
combined = transaction_counts.merge(total_transactions, on='Beneficiary Name')
print(combined)
```

```python
# Filter for success transactions first (adjust column name and value if needed)
success_data = data[data['Transaction Status'] == 'Success']

# Then group by Beneficiary Name
summary = success_data.groupby('Beneficiary Name').agg(
    Count=('Amount', 'size'),
    TotalAmount=('Amount', 'sum')
).reset_index()

print(summary)


```

```python
# Example beneficiary name
name_to_filter = "Swastika Bhusal"

# Filter data for that beneficiary and successful transactions
filtered_data = data[
    (data['Beneficiary Name'] == name_to_filter) & 
    (data['Transaction Status'] == 'Success')
]

# Aggregate count and sum
summary = filtered_data.agg(
    Count=('Amount', 'size'),
    TotalAmount=('Amount', 'sum')
)

print(f"Summary for {name_to_filter} (Success transactions):")
print(summary)
```

```python
# Filter for Swastika Bhusal
prime_data = data[data['Beneficiary Name'] == 'PRIME MOTORS PVT.LTD']

# Group by Amount and count transactions
summary = prime_data.groupby('Amount').size().reset_index(name='Count')

print(summary)
```
