```python
import pandas as pd
import numpy as np
```

```python
data = {
    'name': ['Rob', 'Rafiq', 'Nina','Sofia','Mohan','Tao','Elon Musk'],
    'income': [5000,6000,4000,7500,8000,7000,10000000]
    
}
df = pd.DataFrame(data)
print(df)
```

```python
df.describe()
```

```python
df.income.quantile(0.75)
```

```python
df.income.quantile(0.75, interpolation ='higher')
```

# Remove outlier which is elon musk

```python
percentile_99 = df.income.quantile(0.99)
percentile_99
```

```python
df_no_outlier = df[df.income <=percentile_99]
df_no_outlier
```
