#  Data Cleaning & Visualization 

**Data cleaning** is the process of fixing or removing incorrect, missing, or inconsistent data to make it accurate and usable. It ensures your analysis and models are based on clean, reliable information.

Raw data often contains:

* Missing values
* Duplicates
* Inconsistent formatting
* Outliers
  
We clean data to prepare it for analysis or machine learning.


 # Basic Cleaning with Pandas

<b>Importing libaries</b>

```python
import pandas as pd
import numpy as np
```

# load data

```python

data = pd.read_csv("Titanic-Dataset.csv")
```

# Show first 5 rows

```python
data.head()
```

#  Check for missing values

```python
data.info()
```

```python
data.isnull().sum()
```

# Filling missing values


```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Cabin'] = data['Cabin'].fillna('Unknown')
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])

```

```python
data.isnull().sum()
```

 Now all values are set

# Removing Dublicates if there is any

```python
data.drop_duplicates()
```

<b> so from above we dont need "parch" column so we are droping it</b>

```python

data = data.drop('Parch', axis=1)
```

```python
data
```

# Data Visualization
**Data visualization** is the graphical representation of data and information using charts, graphs, and plots.

It helps you **understand patterns, trends, and outliers** in your data quickly and effectively.


We use Matplotlib and Seaborn to:

* Understand data patterns
* Explore trends and relationships
* Make insights easy to share

<B> Bar Chart: Survived vs Not Survived</b>




```python
import matplotlib.pyplot as plt
data['Survived'].value_counts().plot(kind='bar', color=['red', 'green'])
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.show()
```

# Histogram: Age Distribution

```python
data['Age'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# Pie Chart: Sex Distribution

data['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()


# Seaborn

```python
import seaborn as sns
```

```python
sns.countplot(x='Survived', data=data, palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=data, palette='muted')
plt.title('Survival Count by Gender')
plt.show()
```
