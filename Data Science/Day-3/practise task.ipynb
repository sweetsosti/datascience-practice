```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load mushroom dataset
data = pd.read_csv('mushrooms.csv')

# Encode categorical columns
encoded = data.apply(LabelEncoder().fit_transform)

# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(
    encoded['cap-shape'],
    encoded['stalk-shape'],
    c=encoded['class'], cmap='coolwarm', alpha=0.7
)
plt.title('Cap Shape vs Stalk Shape')
plt.xlabel('Cap Shape')
plt.ylabel('Stalk Shape')
plt.colorbar(label='Class (0 = Edible, 1 = Poisonous)')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='class', y='odor', data=encoded, palette='Set2')
plt.title('Box Plot: Odor by Class')
plt.xlabel('Class (0 = Edible, 1 = Poisonous)')
plt.ylabel('Odor')
plt.show()
```
