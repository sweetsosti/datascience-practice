# Numpy and Pandas

# 1. What is numpy?

 **NumPy** (Numerical Python) is a popular Python library used for numerical computations. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently. NumPy is widely used in data science, machine learning, and scientific computing because it speeds up calculations and makes handling large datasets easier.

Installing & Updating the Package

```python
import numpy as np
```

# example: To find the square of 1-5 using numpy
<b>1. using numpy</b>

```python

number = np.array([1,2,3,4,5])
print(number)
print(number**2)
```

<b> 2. Using list </b>

```python
numbers = {1,2,3,4,5}
for num in numbers:
    print(num ** 2)
```

# 2. What is pandas?
**Pandas** is a powerful Python library used for data manipulation and analysis. It provides easy-to-use data structures like **DataFrame** and **Series** that help organize, clean, and analyze structured data (like tables or spreadsheets). Pandas is essential for tasks such as filtering data, handling missing values, merging datasets, and performing statistical operations, making it a favorite tool for data scientists and analysts.


# Pandas Basic

Installing & Updating the Package Pandas comes pre-isntalled if you're using Anaconda. <b> !pip install pandas</b>

```python
import pandas as pd

data = {
    'Name': ['Shyam', 'Ramesh', 'Sita'],
    'Age': [24, 30, 22],
    'City': ['Kathmandu', 'Pokhara', 'Biratnagar']
}

df = pd.DataFrame(data)
print(df)
```

<b> View first few rows:</b>


```python
df.head()
```

<b>Select a column:</b>

```python
print(df['Age'])
```

<b> Filter rows: </b>

```python
print(df[df['Age'] > 23])
```

<b> add new column </b>

```python
df['Age Next Year'] = df['Age'] + 1
print(df)
```
