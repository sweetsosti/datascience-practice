# 3. Data Structures

 <b> 3.1 List</b>

Lists hold multiple items in a single variable:

```python
fruits = ["apple", "banana", "mango"]
print(fruits[0])  # apple
```

<b> 3.2 Tuples </b>

Tuples are like lists but can’t be changed:

```python
fruits = ("apple", "banana", "mango")
```

<b>3.3  Dictionaries </b>

Used to store data in key-value pairs:

```python
student = {"name": "Sita", "age": 24, "course": "Data Science"}
print(student["name"])  # Swastika
```

# 4. Math in python

In Python, **`math`** is a built-in module that provides functions for performing advanced mathematical operations like square roots, trigonometry, logarithms, and more.


```python
x = 10
y = 3

print(x + y)   # 13
print(x * y)   # 30
print(x / y)   # 3.33
```

# 5. if-else conditions

if-else is used to make decisions in Python based on conditions.
    
<b> for example:</b>

 if condition:
 
     code if condition is True
    
elif another_condition:

    code if another_condition is True
    
else:

    code if none of the above are True




```python
score = 85

if score >= 90:
    print("Excellent")
elif score >= 75:
    print("Good")
else:
    print("Try harder next time!")
```

# 6. Loops

Loops let you repeat a block of code multiple times without writing it again and again.

<b> 6.1 For loops </b>

repeats code for each item in a sequence (like a list or range).

```python
# for loop
for i in range(3):
    print(i)
# Output: 0 1 2
```

<b> 6.2 While loops </b>

repeats code as long as a condition is True.

```python
count = 0
while count < 3:
    print(count)
    count += 1
# Output: 0 1 2
```

# 7. Functions 

A function is a reusable block of code that performs a specific task and can be called anytime in your program.



```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)  # Output: 8
```
